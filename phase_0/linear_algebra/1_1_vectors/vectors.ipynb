## Part 1：向量的三種觀點

不同領域的人對「向量」有不同的直覺：

| 觀點 | 怎麼看向量 | 例子 |
|------|-----------|------|
| **物理學家** | 空間中一個有**方向**和**大小**的箭頭。起點不重要，可以平移。 | 力、速度、位移 |
| **電腦科學家** | 一串**有序的數字**（有序數列）。長度就是維度。 | `[3, 1]` 是 2D 向量 |
| **數學家** | 任何滿足**向量加法**和**純量乘法**的抽象物件。 | 函數、多項式也可以是向量！ |

### 3Blue1Brown 建議的觀點

> 把向量想成**從原點出發的箭頭**。每個向量的**座標**就是它的「身份證」——告訴你怎麼從原點走到箭頭的尖端。

例如，向量 $\vec{v} = \begin{bmatrix} 2 \\ 3 \end{bmatrix}$ 意思是：
- 沿 $x$ 軸走 **2** 步
- 沿 $y$ 軸走 **3** 步

這樣我們就同時擁有了**幾何直覺**（箭頭）和**數值工具**（座標），兩全其美。

## Part 2：用 NumPy 建立向量

先匯入需要的套件，並定義畫圖用的 helper functions。

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- Helper functions ----------

def setup_ax(ax, xlim=(-1, 5), ylim=(-1, 5), title=""):
    """Configure a clean 2D coordinate axis."""
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_aspect("equal")
    ax.grid(True, linewidth=0.3)
    ax.axhline(0, color="black", linewidth=0.5)
    ax.axvline(0, color="black", linewidth=0.5)
    if title:
        ax.set_title(title)
    return ax


def draw_vec(ax, v, color="steelblue", label="", origin=(0, 0)):
    """Draw a 2D vector as an arrow from `origin` to `origin + v`."""
    ax.annotate(
        "",
        xy=(origin[0] + v[0], origin[1] + v[1]),
        xytext=origin,
        arrowprops=dict(arrowstyle="->", color=color, lw=2),
    )
    # Put the label at the midpoint of the arrow
    mid_x = origin[0] + v[0] / 2
    mid_y = origin[1] + v[1] / 2
    if label:
        ax.text(mid_x + 0.15, mid_y + 0.15, label, color=color, fontsize=12)

print("Helpers loaded ✓")
```

```python
# --- Create 2D and 3D vectors ---
v2d = np.array([2, 3])       # 2D vector
v3d = np.array([1, 2, 4])    # 3D vector

print(f"2D vector: {v2d}  (shape: {v2d.shape})")
print(f"3D vector: {v3d}  (shape: {v3d.shape})")
```

```python
# --- Visualize a single 2D vector ---
fig, ax = plt.subplots(figsize=(4, 4))
setup_ax(ax, xlim=(-1, 4), ylim=(-1, 4), title="向量 v = [2, 3]")
draw_vec(ax, v2d, color="steelblue", label="v")
ax.plot(*v2d, "o", color="steelblue", markersize=4)  # dot at tip
plt.show()
```

## Part 3：向量加法

### 理論

兩個向量相加的幾何意義：**尾巴接頭（tip-to-tail）**。

$$\vec{a} + \vec{b} = \begin{bmatrix} a_1 + b_1 \\ a_2 + b_2 \end{bmatrix}$$

步驟：
1. 先畫 $\vec{a}$（從原點出發）
2. 把 $\vec{b}$ 的**尾巴**接到 $\vec{a}$ 的**頭**
3. 從原點到 $\vec{b}$ 的頭，就是 $\vec{a} + \vec{b}$

直覺：想像你先沿 $\vec{a}$ 走一段路，再沿 $\vec{b}$ 走一段路，你最終到達的位置就是 $\vec{a}+\vec{b}$。

```python
# --- Vector addition: tip-to-tail ---
a = np.array([2, 1])
b = np.array([1, 3])
c = a + b  # element-wise addition

print(f"a     = {a}")
print(f"b     = {b}")
print(f"a + b = {c}")

fig, ax = plt.subplots(figsize=(5, 5))
setup_ax(ax, xlim=(-1, 5), ylim=(-1, 5), title="向量加法 (tip-to-tail)")

# Draw a from origin
draw_vec(ax, a, color="steelblue", label="a")
# Draw b from origin (faded) for reference
draw_vec(ax, b, color="darkorange", label="b")
# Draw b shifted to tip of a (tip-to-tail)
draw_vec(ax, b, color="darkorange", label="b′", origin=tuple(a))
# Draw the dashed parallelogram helper line: a shifted to tip of b
draw_vec(ax, a, color="steelblue", label="", origin=tuple(b))
ax.plot([a[0], c[0]], [a[1], c[1]], "--", color="darkorange", linewidth=0.8, alpha=0.5)
ax.plot([b[0], c[0]], [b[1], c[1]], "--", color="steelblue", linewidth=0.8, alpha=0.5)
# Draw resultant a+b
draw_vec(ax, c, color="crimson", label="a+b")

plt.show()
```

## Part 4：純量乘法（Scalar Multiplication）

### 理論

一個純量（scalar）$c$ 乘以一個向量 $\vec{v}$，就是把向量**縮放**：

$$c \cdot \vec{v} = \begin{bmatrix} c \cdot v_1 \\ c \cdot v_2 \end{bmatrix}$$

| 情況 | 效果 |
|------|------|
| $c > 1$ | 向量變**長**，方向不變 |
| $0 < c < 1$ | 向量變**短**，方向不變 |
| $c = 0$ | 向量被壓成**原點** |
| $c < 0$ | 向量**反向**並縮放 |

這就是為什麼我們叫它「**純量**」（scalar）——它的作用就是 **scale**（縮放）向量。

```python
# --- Scalar multiplication: side-by-side comparison ---
v = np.array([1, 2])
scalars = [2, -1, 0.5]
colors = ["steelblue", "crimson", "seagreen"]

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for ax, s, col in zip(axes, scalars, colors):
    setup_ax(ax, xlim=(-3, 4), ylim=(-5, 5), title=f"{s} · v")
    # Draw original vector (faded)
    draw_vec(ax, v, color="gray", label="v")
    # Draw scaled vector
    sv = s * v
    draw_vec(ax, sv, color=col, label=f"{s}v")
    ax.plot(*sv, "o", color=col, markersize=4)

plt.tight_layout()
plt.show()

# Numerical verification
for s in scalars:
    print(f"{s:>4} * {v} = {s * v}")
```

```python
# ===== EXERCISE 1: Vector Addition =====
# ✏️ Change these vectors and re-run!
p = np.array([3, 1])
q = np.array([-1, 2])

result = p + q
print(f"p     = {p}")
print(f"q     = {q}")
print(f"p + q = {result}")

fig, ax = plt.subplots(figsize=(5, 5))
setup_ax(ax, xlim=(-3, 5), ylim=(-2, 5), title="練習：向量加法")
draw_vec(ax, p, color="steelblue", label="p")
draw_vec(ax, q, color="darkorange", label="q")
draw_vec(ax, q, color="darkorange", label="q′", origin=tuple(p))
draw_vec(ax, result, color="crimson", label="p+q")
plt.show()
```
